```python
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
import pandas as pd
```

```python
sentences = ['سلامتی','ورزش','تکنولوژی','سالم','غذای سالم خوردن','شب غذا سبک بخور','با همدیگر ورزش کنیم','گلف','برنامه نویسی','مصاحبه مهندسی']             
# print(nlp.predict(sentences[0]).to_numpy()[0][1])
```

```python
#@title Default title text
# Turn into vectors
vectors = [nlp.predict(sentence).to_numpy()[0][1] for sentence in sentences]
vector=[]
#for i in range(len(vectors)):
#  vector.append(vectors[i]['word_embedding_word2vec'])
#type(vector) 
# import array
  
# # sample_array = array.fromlist(vector)  
# sample_array = np.array(vectors)
# sample_array
# Compute similarities
similarities = cosine_similarity(vectors)

similarity_df = pd.DataFrame(similarities,
                             index=sentences,
                             columns=sentences) 
                                                         
similarity_df
```

```python
room_list = similarity_df.columns.values
# interest
interests = ['تکنولوژی' ,'ورزش' ,'سلامتی']

top_n = 3
for i in range(0,len(interests)):
   room_sim = similarity_df[similarity_df.index == interests[i]].values[0] 
   sorted_room_ids = np.argsort(room_sim)[::-1] 
   recommended_rooms = room_list[sorted_room_ids[1:top_n+1]]
   print('\n\nTop Recommended room for:', interests[i], 'are:-\n', recommended_rooms)
```
